```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
data = pd.read_csv('C:\\Users\\Dell\\Downloads\\test.csv')
```

```python
display(data.head(10))
```

```python
#scatter plot battery_power against fc
plt.scatter(data['battery_power'], data['fc'])

#adding title to the plot 
plt.title("scatter plot")

#setting the x and y Labels 
plt.xlabel('battery_power')
plt.ylabel('fc')

#adding the legends
plt.show()
```

```python
#line chart with year against own
plt.plot(data['battery_power'])
plt.plot(data['fc'])

#adding title to the plot 
plt.title("Line chart")

#setting the x and y labels
plt.xlabel('batery_power')
plt.ylabel('fc')

#adding the legends
plt.show()
```

```python
#Bar chart or bar plot
plt.bar(data['battery_power'], data['fc'])

#adding title to the plot
plt.title("bar Chart")

#setting the x and y Lables 
plt.xlabel('battery_power')
plt.ylabel('fc')

#adding the legends
plt.show()
```
